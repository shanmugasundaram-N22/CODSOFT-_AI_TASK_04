```python
app_code = '''
import streamlit as st
from PIL import Image
import torch
import json

# Load YOLOv5 model
model = torch.hub.load('yolov5', 'custom', path='best.pt', source='local')

# Load care guide
with open('care_guides.json', 'r') as f:
    care_data = json.load(f)

st.title("ðŸŒ³ Bonsai Species Classifier & Care Recommendation")
st.markdown("Upload a bonsai image to detect the species and get care tips.")

uploaded_file = st.file_uploader("Choose a bonsai image...", type=['jpg', 'jpeg', 'png'])

if uploaded_file:
    img = Image.open(uploaded_file)
    st.image(img, caption='Uploaded Image', use_column_width=True)

    with st.spinner("Detecting species..."):
        results = model(img)
        detections = results.pandas().xyxy[0]['name'].tolist()

    if detections:
        st.success(f"Detected: {', '.join(set(detections))}")
        for species in set(detections):
            st.subheader(f"ðŸª´ Species: {species}")
            if species in care_data:
                tips = care_data[species]
                st.markdown("**ðŸŒ± Care Tips:**")
                st.markdown(f"- **Watering**: {tips['watering']}")
                st.markdown(f"- **Sunlight**: {tips['sunlight']}")
                st.markdown(f"- **Fertilizer**: {tips['fertilizer']}")
                st.markdown(f"- **Pruning**: {tips['pruning']}")
            else:
                st.warning("Care data not found for this species.")
    else:
        st.error("No bonsai detected in the image.")
'''

with open('app.py', 'w') as f:
    f.write(app_code)
```
